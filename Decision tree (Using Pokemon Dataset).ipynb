```python
#now extracting dependent and independent varible

# x=speed(independent)
#y=has mega evolution(dependent)
```

```python
x=Poke[['Speed']]
```

```python
y=Poke[['hasMegaEvolution']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#now making decision tree classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc= DecisionTreeClassifier()
```

```python
#now fits DecisionTreeClassifier on X_train and Y_train
```

```python
dtc.fit(x_train,y_train)
```

```python
#predict the results
```

```python
y_pred= dtc.predict(x_test)
```

```python
#checking diffirence between actual value and predicted value
#import confussion matrix
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```
